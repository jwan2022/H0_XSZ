```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
other_h0_values = [67.3, 70.0, 66.98, 67.4, 74.03, 69.8, 73.3, 69.8]
errors = [[1.1, 8.0, 1.18, 0.5, 1.42, 0.8, 1.8, 15.4], [1.1, 12.0, 1.18, 0.5, 1.42, 0.8, 1.7, 15.4]]
y = [1, 2, 3, 4, 5, 6, 7, 8]
names = ['Aubourg et al. 2015 (BAO + SNe Ia)', 'Abbott et al. 2017 (GW)', 'Addison et al. 2018 (BAO + D/H)', 
         'Planck Collaboration 2018 (CMB)', 'Riess et al. 2019 (SNe Ia)', 'Freedman et al. 2019 (TRGB)',
         'Wong et al. 2020 (Gravitational lensing)', 'This work']

fig, ax = plt.subplots(figsize=(11,6))
plt.ylim(top=9)

for i in range(0, len(y)):
    plt.errorbar(other_h0_values[i], y[i], xerr=[[errors[0][i]], [errors[1][i]]], linestyle='', marker='o', 
                 capsize=5, label=names[i], markersize=8, linewidth=2.5, color = 'C' + str(i+4))
    plt.annotate(names[i], (other_h0_values[i], y[i]+0.3), ha='center')
plt.errorbar(69.8, 6, xerr=[[0.8+1.7], [0.8+1.7]], linestyle='', marker='o', capsize=5, linewidth=1, color='C' + str(5+4))
#plt.legend()
plt.xlabel('$H_0$ (km s$^{-1}$ Mpc$^{-1}$)', fontsize=14)
plt.yticks([], [])
ax.tick_params(direction='in', length=14)
ax.tick_params(which='minor', direction='in', length=6)
ax.xaxis.set_minor_locator(MultipleLocator(2.5))
#plt.savefig('/Users/jwan/Documents/research 2020/plots for paper/H0_comparison_general_new.png')
plt.show()
```
